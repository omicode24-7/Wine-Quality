```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.metrics import confusion_matrix,classification_report,f1_score,precision_score,recall_score,mean_squared_error,r2_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC,SVC
```

```python
X=df_new.drop('quality',axis=1)
Y=df_new['quality']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler()
X_train1,Y_train1=ros.fit_resample(X_train,Y_train)
X_test1,Y_test1=ros.fit_resample(X_test,Y_test)
```

```python
def create_model(model):
    model.fit(X_train,Y_train)
    Y_pred=model.predict(X_test)
    print(classification_report(Y_test,Y_pred))
    print(confusion_matrix(Y_test,Y_pred))
```

```python
def regression(model):
    model.fit(X_train1,Y_train1)
    Y_pred=model.predict(X_test1)
    print("MSE:",mean_squared_error(Y_test1,Y_pred))
    print("r2 score",r2_score(Y_test1,Y_pred))
```

```python
lr=LinearRegression()
model=regression(lr)
```

```python
dt1=DecisionTreeClassifier()
model=regression(dt1)
```

```python
dt1=DecisionTreeClassifier(max_depth=6)
model=regression(dt1)
```

```python
# Forward Selection
X=df_new.drop('quality',axis=1)
Y=df_new['quality']
```

```python
columns=[]
for col in X:
    columns.append(col)
    X_new=df_new[columns]
    X_train,X_test,Y_train,Y_test=train_test_split(X_new,Y,test_size=0.3,random_state=1)
    X_train1,Y_train1=ros.fit_resample(X_train,Y_train)
    X_test1,Y_test1=ros.fit_resample(X_test,Y_test)
    lr=LinearRegression()
    lr.fit(X_train1,Y_train1)
    score=lr.score(X_test1,Y_test1)
    print(columns,":",np.round(score,3))
```

```python
# Backward Selection Method
X=df_new.drop('quality',axis=1)
Y=df_new['quality']
```

```python
columns=[]
X_new=X
n_col=X_new.shape[1] #no. of columns
for i in range(n_col,0,-1): 
    columns.append(X_new)
    X_train,X_test,Y_train,Y_test = train_test_split(X_new,Y,test_size=0.3,random_state=1)
     #create a object of LinearRegression
    X_train1,Y_train1=ros.fit_resample(X_train,Y_train)
    X_test1,Y_test1=ros.fit_resample(X_test,Y_test)
    lr=LinearRegression()
    #we train the model
    lr.fit(X_train1,Y_train1)
    #find the score
    score1=lr.score(X_test1,Y_test1)
    print(i,":",np.round(score1,3))
    X_new=X_new.iloc[:,:-1]  #iloc index location iloc[startrow:stoprow ,startcol:-1]
    #iloc[rowindex,colindex]
    print("After Remove Column : ", i)
```
